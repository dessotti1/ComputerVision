Transformações geométricas - ROTAÇÃO, TRANSLACAO e AJUSTE DE ESCALA

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

Ler imagem

```python
imgOriginal = cv.imread("einstein.jpg", 0)
```

ROTAÇÃO

Função getRotationMatrix2D e warpAffine

```python
linhas, colunas = imgOriginal.shape
print(linhas, colunas)
```

```python
mat = cv.getRotationMatrix2D((linhas/2, colunas/2), 45, 1)
```

```python
imgRotacionada = cv.warpAffine(imgOriginal, mat, (linhas, colunas))
```

Plotar

```python
fig = plt.figure(figsize=(10,8))

ax1 = fig.add_subplot(121)
plt.imshow(imgOriginal, cmap=plt.cm.gray)

ax2 = fig.add_subplot(122)
plt.imshow(imgRotacionada, cmap=plt.cm.gray)

plt.show()
```

TRANSLAÇÃO

Função float32

```python
mat = np.float32([[1,0,100], [0,1,100]])
imgDeslocada = cv.warpAffine(imgOriginal, mat, (linhas, colunas))
```

Plot

```python
fig = plt.figure(figsize=(10,8))

ax1 = fig.add_subplot(121)
plt.imshow(imgOriginal, cmap=plt.cm.gray)

ax2 = fig.add_subplot(122)
plt.imshow(imgDeslocada, cmap=plt.cm.gray)

plt.show()
```

AJUSTE DE ESCALA

Função resize

```python
imgModificada = cv.resize(imgOriginal, None, fx=0.8, fy=0.8, interpolation = cv.INTER_CUBIC)
```

```python
fig = plt.figure(figsize=(10,8))

ax1 = fig.add_subplot(121)
plt.imshow(imgOriginal, cmap=plt.cm.gray)
plt.title(imgOriginal.shape)

ax2 = fig.add_subplot(122)
plt.imshow(imgModificada, cmap=plt.cm.gray)
plt.title(imgModificada.shape)

plt.show()
```
