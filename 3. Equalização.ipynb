Equalização do histograma

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

Ler a imagem

```python
img = cv.imread("einstein.jpg", 0)
```

Função equalizeHist

```python
imgEq = cv.equalizeHist(img)
```

```python
fig = plt.figure(figsize=(10,5))

ax1 = fig.add_subplot(121)
plt.imshow(img, cmap=plt.cm.gray)

ax2 = fig.add_subplot(122)
plt.imshow(imgEq, cmap=plt.cm.gray)

plt.show()
```

Histogramas

```python
fig = plt.figure(figsize=(10,5))

ax1 = fig.add_subplot(121)
ax1.hist(img.ravel(), 256, [0,256])
plt.title("Histograma da imagem original")

ax2 = fig.add_subplot(122)
ax2.hist(imgEq.ravel(), 256, [0,256])
plt.title("Histograma da imagem equalizada")

plt.show()
```
