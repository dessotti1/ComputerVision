```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

Ler imagens

```python
img1 = cv.imread("superman.png", 0)
img2 = cv.imread("batman.png", 0)
```

Mostrar as imagens

```python
fig = plt.figure(figsize=(10,8))

ax1 = fig.add_subplot(121)
plt.imshow(img1, cmap=plt.cm.gray)

ax2 = fig.add_subplot(122)
plt.imshow(img2, cmap=plt.cm.gray)

plt.show()
```

Função add e subtract


```python
imgSum = cv.add(img1, img2)
imgSub = cv.subtract(img1, img2)
```

Plot

```python
fig = plt.figure(figsize=(10,8))

ax1 = fig.add_subplot(121)
plt.imshow(imgSum, cmap=plt.cm.gray)

ax2 = fig.add_subplot(122)
plt.imshow(imgSub, cmap=plt.cm.gray)

plt.show()
```

Mistura

Função addWeighted

```python
imgMisturada = cv.addWeighted(img1, 0.2, img2, 0.8, 0)
plt.imshow(imgMisturada, cmap=plt.cm.gray)
```
