FILTRO MÉDIA

 * substitui cada pixel da imagem pelo valor médio de sua vizinhança

```python
import cv2 as cv
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv.imread("einstein.jpg")
```

Função blur

```python
imgM = cv.blur(img, (5, 5))
```

```python
plt.figure(figsize=(10,8))
plt.subplot(1, 2, 1)
plt.imshow(img)

plt.subplot(1, 2, 2)
plt.imshow(imgM)

plt.show()
```
