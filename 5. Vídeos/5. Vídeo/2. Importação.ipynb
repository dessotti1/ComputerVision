Importação do vídeo

```python
import cv2
import time
from google.colab.patches import cv2_imshow

cap = cv2.VideoCapture('teste.mp4')

if cap.isOpened() == False:
  print('Error')
```

Rodar o vídeo

```python
while cap.isOpened():
  ret, frame = cap.read()

  if ret == True:
    time.sleep(1/20)
    cv2_imshow(frame)

    if cv2.waitKey(10) & 0xFF == ord('q'):
      break
  
  else:
    break

cap.release()
cv2.destroyAllWindows()
```
