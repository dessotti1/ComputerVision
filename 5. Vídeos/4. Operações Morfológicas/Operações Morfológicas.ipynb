É uma transformação na imagem, a qual é realizada usando uma segunda imagem menor de estrutura e geometria conhecidas, denominada elemento estruturante.

* utilizado para ressaltar ou remover aspectos específicos dos objetos da imagem maior.

A biblioteca cv2 tem três elementos estruturantes básicos (retangular,
elíptico e cruz) e, para usá-los, é preciso usar o comando getStructuringElement.

- RETANGULAR: cv2.MORPH_RECT;
- ELÍPTICO: cv2.MORPH_ELLIPSE;
- CRUZ: cv2.MORPH_CROSS.

EROSÃO: corrosão das arestas do objeto de interesse, “encolhendo” a
imagem do objeto.

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread("placa.PNG")
```

```python
elem_estr = cv.getStructuringElement(cv.MORPH_RECT, (7, 7))
print(elem_estr)
```

```python
imgE = cv.erode(img, elem_estr, iterations = 2)

plt.figure(figsize=(10,8))
plt.subplot(2, 1, 1)
plt.imshow(img)

plt.subplot(2, 1, 2)
plt.imshow(imgE)

plt.show()
```

DILATAÇÃO: é o oposto da erosão. Nesse processo, a área do objeto de
interesse será dilatada, ou seja, ficará maior.

```python
elem_estr2 = cv.getStructuringElement(cv.MORPH_ELLIPSE, (7, 7))
print(elem_estr)
```

```python
imgD = cv.dilate(img, elem_estr2, iterations = 2)

plt.figure(figsize=(10,8))
plt.subplot(2, 1, 1)
plt.imshow(img)

plt.subplot(2, 1, 2)
plt.imshow(imgD)

plt.show()
```

OUTROS

- GRADIENTE MORFOLÓGICO: diferença entre dilatação e erosão. O resultado é uma imagem que representa a borda do objeto;

- TOPHAT: diferença entre a imagem de entrada e a abertura da imagem. É aplicado para realçar objetos brilhantes em fundos escuros e para corrigir a variação de luminosidade em imagens;

- BLACKHAT: diferença entre o fechamento da imagem e a imagem de
entrada.


Função morphologyEx

Métodos:
* cv2.MORPH_GRADIENT;
* cv2.MORPH_TOPHAT;
* cv2.MORPH_BLACKHAT.

```python
img1 = cv.imread("moeda.PNG")
img2 = cv.imread("esqueleto.tif")

plt.figure(figsize=(15,15))
plt.subplot(1, 2, 1)
plt.imshow(img1)

plt.subplot(1, 2, 2)
plt.imshow(img2)

plt.show()
```

```python
elem_str1 = cv.getStructuringElement(cv.MORPH_ELLIPSE, (5,5))
elem_str2 = cv.getStructuringElement(cv.MORPH_ELLIPSE, (21,21))

print(elem_str1)
print(elem_str2)
```

```python
imgProc1 = cv.morphologyEx(img1, cv.MORPH_GRADIENT, elem_str1)
imgProc2 = cv.morphologyEx(img2, cv.MORPH_TOPHAT, elem_str2)
imgProc3 = cv.morphologyEx(img2, cv.MORPH_BLACKHAT, elem_str2)

plt.figure(figsize=(15,15))
plt.subplot(1, 3, 1)
plt.imshow(imgProc1)
plt.title("Gradiete Morfologico")

plt.subplot(1, 3, 2)
plt.imshow(imgProc2)
plt.title("TOPHAT")

plt.subplot(1, 3, 3)
plt.imshow(imgProc3)
plt.title("BLACKHAT")

plt.show()
```
