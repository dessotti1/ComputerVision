FILTRO DE SOBEL

* realiza uma operação para realçar contornos em imagens. 

* não linear 

* realça as linhas verticais e horizontais mais escuras que o fundo, sem realçar pontos isolados.

* possibilita realçar as arestas independente da direção: a filtragem
funciona vertical e horizontalmente. As regiões destacadas por esse filtro
resultam em bordas mais grossas.

```python
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
img = cv.imread("predio.JPG")
```

```python
sobelx = cv.Sobel(img, cv.CV_8U, 0, 1, ksize=7)
sobely = cv.Sobel(img, cv.CV_8U, 1, 0, ksize=7)
```

```python
plt.figure(figsize=(20,50))

plt.subplot(1, 3, 1)
plt.imshow(img)

plt.subplot(1, 3, 2)
plt.imshow(sobelx)

plt.subplot(1, 3, 3)
plt.imshow(sobely)

plt.show()
```
