FILTRO DE CANNY

* um dos algoritmos mais eficientes para detectar bordas em imagens
* deve ser capaz de identificar todas as bordas possíveis
* todas as bordas detectadas devem estar próximas das bordas originais 
* bordas falsas não podem ser criadas.

PROCESSO:

i. Filtragem gaussiana: imagem suavizada pelo filtro gaussiano;

ii. Detecção de intensidade de gradientes: operadores de gradientes;

iii. Supressão de pixels não máximos: faz o afinamento das bordas;

iv. Duplo limiar: cada pixel é comparado com dois limiares e, então, é
classificado.

v. Rastreamento de bordas: verificar se os pixels de bordas fracas
estão conectados com os de bordas fortes.

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread("predio.JPG")
```

```python
imgC = cv.Canny(img, 75, 50)

plt.figure(figsize=(20,50))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap=plt.cm.gray)
plt.title("Imagem Original")

plt.subplot(1, 2, 2)
plt.imshow(imgC, cmap=plt.cm.gray)
plt.title("Operador Canny")

plt.show()
```
