FILTRO LAPLACIANO

* máscara de ordem 3 que percorre toda a imagem
alterando o pixel-alvo pela média ponderada dos pixels vizinhos, e depois
eleva ao quadrado obtido.

* produz bordas mais finas que o Sobel, mas é
muito sensível a ruídos.

```python
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
img = cv.imread("predio.JPG")
```

```python
imgL = cv.Laplacian(img, cv.CV_8U)

plt.figure(figsize=(20, 50))
plt.subplot(1, 2, 1)
plt.imshow(img)

plt.subplot(1, 2, 2)
plt.imshow(imgL)

plt.show()
```
