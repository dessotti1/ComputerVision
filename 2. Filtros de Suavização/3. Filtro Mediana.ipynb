FILTRO MEDIANA
* não linear
* trata ruídos do tipo sal e pimenta

```python
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
img = cv.imread("head.png")
```

```python
imgMed = cv.medianBlur(img, 5)

plt.figure(figsize=(20,50))
plt.subplot(1, 2, 1)
plt.imshow(img)

plt.subplot(1, 2, 2)
plt.imshow(imgMed)
plt.show()
```
